```python
# income attribute set up
bank_income = bank_file['income']
bank_income.head()
```

```python
# Z-score for income
income_zscore = (bank_income-bank_income.mean())/bank_income.std()
income_zscore.head()
```

```python
# Decritize Age attribute to 3 categories (young, middle-age , old)
```

```python
bank_file['age'].describe()
```

```python
#Z-score Age
age_z = (bank_file['age']-bank_file['age'].mean()) / bank_file['age'].std() 
age_z.head()
```

```python
#Z-score Age income children
IAG = bank_file[['income','age','children']]
z_IAG = (IAG-IAG.mean()) / IAG.std()
z_IAG.rename(columns={'income':'income_z','age':'age_z','children':'children_z'},inplace = True)
z_IAG.head()
```

```python
#Adding new column 'Age-std'
bank_file['Age-std'] = age_z
bank_file.head()
```

```python
Age_bins = pd.qcut(bank_file.age, 3)
Age_bins.head(10)
```

```python
# use qcut() to discretize age
age = pd.qcut(bank_file.age,[0,.33,.66,1],labels = ['young','middle-age','old'])
age
```

```python
# Add variable age into the dataframe by using pd.concat()
bank_file2 = pd.concat([bank_file,age], axis = 1)
bank_file2.head()
```

```python
# Add variable age into dataframe bank_file
bank_file['Age-c'] = age
bank_file.head()
```

```python
# Set min and max for income , age, children
min_n =bank_file[['income','age','children']].min()
max_n =bank_file[['income','age','children']].max()
print(min_n)
print(max_n)
```

```python
## Normalization 0-1
normalize = (bank_file[['income','age','children']] - min_n) / (max_n-min_n)
print(normalize.head(10))
```

```python
normalize.rename(columns = {'income':'income_n','age':'age_n','children':'children_n'}, inplace = True)
normalize.head()
```

```python
# Add variable age into the dataframe by using pd.concat()
bank_file3 = pd.concat([bank_file,normalize,z_IAG], axis = 1)
bank_file3.head()
```

```python
# remove column Age-c
bank_file4=bank_file3.drop(['Age-c','id'],axis=1)
bank_file4.head()
```

```python
# Numeric dataframe including Z-score and Normalization with income age and children.
bank_file4 = pd.get_dummies(bank_file4)
bank_file4.head()
```

```python
bank_file4.to_csv("Desktop/DSC 478/DSC 478 Week 2/Assignment 1/bank_numeric.csv", float_format="%1.2f")
```
